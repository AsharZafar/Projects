#Creating models

```python
X = df.drop(['YIELD', 'AC_POWER', 'DC_POWER', 'date_time',"dated"], axis=1)
X
```

```python
df.head()
```

```python
Y = df['YIELD']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of Y_train: {Y_train.shape}")
```

```python
print(Y_train.isnull().sum())  # Check for NaNs
print(Y_train.value_counts())  # Check for unexpected values
```

```python

# Update your model_param dictionary with regression models
model_param = {
    'random_forest': {
        'model': RandomForestRegressor(),
        'param': {
            'n_estimators': [10, 50, 100],
            'max_depth': [None, 10, 20]
        }
    },
    'svr': {
        'model': SVR(),
        'param': {
            'C': [1, 10, 100],
            'kernel': ['rbf', 'linear']
        }
    }
}
```

```python
scores = []

for model_name, mp in model_param.items():
    clf = GridSearchCV(mp["model"], mp["param"], cv=5, return_train_score=False)
    clf.fit(X_train, Y_train)
    scores.append({
        "model": model_name,
        "best_score": clf.best_score_,
        "best_params": clf.best_params_
    })

# Print the results
print(scores)
```
